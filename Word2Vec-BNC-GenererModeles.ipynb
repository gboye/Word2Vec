### Formes, fréquences, surabondances

```python
def calcFormFreq(verb,form,cell):
    result=[[None,None],[None,None],[None,None]]
    if verb not in verbs:
        result[cell]=[form,1]
    else:
        result=verbs[verb]
        if verbs[verb][cell][0] and verbs[verb][cell][0]==form:
            result[cell]=[verbs[verb][cell][0],verbs[verb][cell][1]+1]
        elif verbs[verb][cell][0]:
#            print ("overabondance",verbs[verb][cell],form)
            if verb not in overabundance:
                overabundance[verb]=[None,None,None]
            if cell in overabundance[verb]:
                if form not in overabundance[verb][cell]:
                    overabundance[verb][cell].append(form)
            else:
                overabundance[verb][cell]=[verbs[verb][cell][0],form]
        else:
            result[cell]=[form,1]
    return result
    
```
